```python
from pyspark.sql import SparkSession
```

```python
spark=SparkSession.builder.appName('dataframe').getOrCreate()
```

```python
df_pyspark =spark.read.csv('TSLA.csv', header=True, inferSchema=True)
df_pyspark.show()
```

### Filter operations

```python
## Open stock of Tesla is less than or equal to 40.00

df_pyspark.filter('Open<=40.00').show()

## All details will be stock open less than or equal to 40.00
```

```python
## To select tow column as per filter

df_pyspark.filter('Open<=40.00').select(['Date','Open','Close']).show()  # We can put one column in list as well
```

```python
# We can also write the filter like below ---

df_pyspark.filter(df_pyspark['Open']<=40.00).show()  # Like pandas


## For specifc columns

df_pyspark.filter(df_pyspark['Open']<=40.00).select(['Date','Open','Close']).show()
```

```python
# To givee multi filter together

df_pyspark.filter((df_pyspark['Open']<=40.00) & 
                  (df_pyspark['Open']>=38.00)).show()
```

```python
## use | for or,  & for and operations in between two or more filter

df_pyspark.filter((df_pyspark['Open']<=40.00) | 
                  (df_pyspark['Open']>=38.00)).show()
```

```python
# ~ ---> for not operations

df_pyspark.filter(~(df_pyspark['Open']<=40.00)).show()
```
